```python
from keras.models import load_model
model = load_model('croquetas_and_tortilla_2.h5')
```

```python
model
```

```python
import keras as keras
```

```python
image = keras.preprocessing.image.load_img('tortilla_test.jpg', target_size=(150, 150))
```

```python
image
```

```python
import numpy as np
input_arr = keras.preprocessing.image.img_to_array(image)
input_arr = np.array([input_arr])  # Convert single image to a batch.
```

```python
input_arr
```

```python
input_arr = input_arr.astype('float32') / 255
```

```python
input_arr
```

```python
predictions = model.predict(input_arr)
```

```python
print(predictions)
```

```python
score = predictions[0]
print(
    "Este plato es %.2f por ciento croquetas y %.2f por ciento tortilla de patatas."
    % (100 * (1 - score), 100 * score)
)
```

```python
image2 = keras.preprocessing.image.load_img('croquetas_test.jpg', target_size=(150, 150))
```

```python
image2
```

```python
input_arr = keras.preprocessing.image.img_to_array(image2)
input_arr = np.array([input_arr])  # Convert single image to a batch.
input_arr = input_arr.astype('float32') / 255
```

```python
predictions = model.predict(input_arr)
```

```python
print(predictions)
```

```python
score = predictions[0]
print(
    "Este plato es %.2f por ciento croquetas y %.2f por ciento tortilla de patatas."
    % (100 * (1 - score), 100 * score)
)
```
